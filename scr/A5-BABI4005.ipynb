```python
# Imports all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Loads dataset #1
data = pd.read_csv("../data/GenZ_DatingApp_Data.csv")
```

```python
data2 = pd.read_csv("../data/DatingData2.csv")
```

```python
# Merge tha data

merged_data = pd.merge(data, data2, on = 'User_ID', how = 'outer')
```

##### Why I Chose an Outer Join
- An outer join ensures that all rows from both datasets are included in the final combined dataset, even if there are no matching rows in the other dataset.
- If there are rows in one dataset that do not have corresponding matches in the other, the outer join will still include them, filling in NaN (Not a Number) for missing values.
- This approach allows us to analyze the entire dataset, including records that may not have matches, which is useful for identifying gaps or inconsistencies.

##### Type of NA Values Created
- If a row exists in clean_data but not in new_data (or vice versa), the columns from the missing dataset will be filled with NaN.
- If the datasets have columns with the same name (other than the join key), pandas will automatically append suffixes (_x for the left dataset and _y for the right dataset) to distinguish them. These columns may also contain NaN values if there are no matches.

```python
# Display the merged dataset

merged_data
```

```python
# Drops rows where all columns are NA values

clean_merge = merged_data.dropna(how = "all")
```
